```
!pip install botnoi==0.2.1
import botnoi as bn
from botnoi import scrape as sc
from botnoi import cv
import os
import urllib.request

def extractimagefeat(query):
  #create folder
  foldername = 'D:/BOTNOI Computer Vision/images/'+query
  isdir = os.path.isdir(foldername) 
  #check if folder exist
  if not isdir:
    #create directory
    os.makedirs(foldername)
  #get images from google search
  imglist = sc.get_image_urls(query, 1000)
  i = 1
  for img in imglist[0:1000]:
    #extract image features from each images and save to files
    try:
      print(i)
      #create image path
      savepath = foldername + '/' + str(i)+'.jpg'
      print(savepath)
      urllib.request.urlretrieve(img, savepath)
      i = i + 1
    except:
      pass
  return 'complete'

clsList = ['หนังสือ', 'เสื้อผ้า', 'ลิปสติก', 'รองเท้า']
for c in clsList:
  extractimagefeat(c)
```
